```python
import torch
```

```python
inputs = torch.nn.Embedding( 4, 8 )
```

```python
inputs = inputs.weight.data
inputs
```

```python
#set dimension
d_in = 8
d_out = 6

#create weight matrices
W_q = torch.nn.Parameter( torch.rand( d_in, d_out ), requires_grad=False )
W_k = torch.nn.Parameter( torch.rand( d_in, d_out ), requires_grad=False )
W_v = torch.nn.Parameter( torch.rand( d_in, d_out ), requires_grad=False )
```

```python
#choose an input vector and transform it into the query vector using W_q
query = inputs[2] @ W_q
print("query:",query) #as you can see the vector is now of length six
```

```python
#calc attention scores using the keys generated by W_k
keys = inputs @ W_k #in that order for both
values = inputs @ W_v
print("keys:", keys)
print("values:", values)
```

```python
attention_scores = query @ keys.T #in that order
attention_scores
```

```python
attention_weights = torch.softmax(attention_scores, dim=-1)
attention_weights
```

```python
attention_weights.sum()
```

```python
context_vector = attention_weights @ values #in that order
context_vector
```
